# Exploratory Data Analysis

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Download Dataset from Kaggle

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("abdallahwagih/books-dataset")

print("Path to dataset files:", path)
```

## Load and View the Dataset

```python
books = pd.read_csv(path + "/data.csv")
books
```

## Check for Missing Values

```python
books.isna().sum()
```

## Visualise the Missing Values to Detect any Patterns

```python
sns.heatmap(books.isna().transpose(), cbar=False)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.title("Missing Values in the Dataset")
plt.show()
```

From the above heatmap, we can observe 

- subtitle has a lot of missing values (but it doesn't affect negatively for our project)
- the missing values in categories, thumbnail and description do not show any pattern
- the missing values in average_rating, num_pages and ratings_count have a pattern
  - they share common books
  - could mean they were collected from different dataset

## Check if missing description is related to any features of book

Since description would be the most important feature for our project, it behooves us to check if the missing description is related to any other features in the book. If there is, and we drop the books with missing description, our model might be bias towards certain books.

Check if there is any relation between description and
- length of books (short books / long books)
- age of books (older books / newer books)

```python
# check if missing description is based on length of book / age of book
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]

columns_of_interest = ["num_pages", "missing_description", "age_of_book"]
```

```python
# calculate correation and visualise in heatmap to find if missing description is based on length of book / age of book
correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

The above heatmap shows no correlation between missing description and length of book or age of book, so we can drop the sample of books with missing description without risking bias from our model.

```python
complete_books = books[~(books["description"].isna()) &
                       ~(books["num_pages"].isna()) & 
                       ~(books["published_year"].isna()) & 
                       ~ (books["average_rating"].isna())
                      ]

complete_books
```

## Check for Non-informative Descriptions

```python
# calculate the number of words in the description to find short descriptions

complete_books['words_in_description'] = complete_books['description'].str.split().str.len()
complete_books.head()
```

```python
# Plot the histogram of words in description
sns.histplot(complete_books['words_in_description'], bins=50, kde=True)
```

```python
# Check the description of books with words between 25 and 34 to see if they are useful
complete_books.loc[complete_books['words_in_description'].between(25, 34), 'description']
```

```python
# Remove books with less than 25 words in the description
complete_books_25_words = complete_books.loc[complete_books['words_in_description'] > 25]
complete_books_25_words
```

```python
# Combine title and subtitle
complete_books_25_words['title_and_subtitle'] = (
  np.where(complete_books_25_words['subtitle'].isna(),
           complete_books_25_words['title'],
           complete_books_25_words[['title', 'subtitle']].astype(str).agg(': '.join, axis=1)
          )
)

complete_books_25_words
```

```python
# Create a tagged description column and save the dataset
complete_books_25_words['tagged_description'] = complete_books_25_words[['isbn13', 'description']].astype(str).agg('_'.join, axis=1)
complete_books_25_words
```

```python
# Remove the additional columns created for analysis and save the dataset
complete_books_25_words.drop(['subtitle', 'words_in_description', 'missing_description', 'age_of_book'], axis=1).to_csv('books_cleaned.csv', index=False)
```
